```python
CONTROL_NUM = 5
FAVORITE_ARTIST = "KESHI"

def audit_log(func):
    def wrapper(power):
        print("Authorization Started.")
        result = func(power)
        print("Authorization Completed.")
        return result
    return wrapper

def signal_shutdown(power):
    print("Current Signal Strength:", power)
    if power == 0:
        return 1
    return 1 + signal_shutdown(power - 1)

@audit_log
def start_shutdown(power):
    return signal_shutdown(power)

power = CONTROL_NUM + len(FAVORITE_ARTIST)

calls = start_shutdown(power)

print("Total Recursive Calls:", calls)
```
